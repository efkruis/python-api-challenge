```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
long_range = (-180, 180)
```

## Generate Cities List

```python
# Lists
lat_longs = []
cities = []

# Random lat and long combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
longs = np.random.uniform(long_range[0], long_range[1], size=1500)
lat_longs = zip(lats, longs)

# Identify nearest city for each lat, lng combination
for lat_long in lat_longs:
    city = citipy.nearest_city(lat_long[0], lat_long[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
cities
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = ["metric", "imperial"]
query_url = f"{url}appid={weather_api_key}&units={units}&q="
```

```python
response = requests.get(query_url + city).json()
response
```

```python
response.keys()
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
weather = []
count = 0

for city in cities:
    weather_api = requests.get(query_url + city).json()
    try:
        weather_dict = {
            "City":weather_api['name'] ,
            "Lat": weather_api['coord']['lat'],
            "Lng": weather_api['coord']['lon'],
            "Max Temp": weather_api['main']['temp_max'],
            "Humidity": weather_api['main']['humidity'],
            "Cloudiness": weather_api['clouds']['all'],
            "Wind Speed": weather_api['wind']['speed'],
            "Country": weather_api['sys']['country'],
            "Date": weather_api['dt']}
        weather.append(weather_dict)
        count = count + 1
        print(f"Processing {count} {weather_api['name']}")
    except:
        print(f"City not found. Skipping..")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_data = pd.DataFrame(weather)
weather_data.head()
```

```python
#exporting data into a csv
weather_data.to_csv("output_data/cities.csv", index = False)
```

#  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = north_hem['Max Temp']
y_values = north_hem['Lat']

# Perform a linear regression on temperature vs. latitude
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

# Get regression values
regress_values = x_values * slope + intercept
```

```python
# Create Plot
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.xlabel('Max Temp')
plt.ylabel('Latitude')
plt.title('Northern Hemisphere - Max Temp vs. Latitude Linear Regression')
```
